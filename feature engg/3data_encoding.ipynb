```python
import seaborn as sns

sns.load_dataset('tips')
```

```python

df=sns.load_dataset('tips')
# df=df.head()
```

```python
# Step 4: Select categorical columns to encode
categorical_cols = ['sex', 'smoker', 'day', 'time']

# Step 5: Initialize OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, drop='first')  # drop='first' to avoid dummy variable trap

# Step 6: Fit and transform the categorical columns
encoded_array = encoder.fit_transform(df[categorical_cols])

# Step 7: Convert the encoded data back to a DataFrame
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(categorical_cols))

# Step 8: Concatenate the encoded columns with the original DataFrame (excluding the original categorical columns)
tips_encoded = pd.concat([df.drop(categorical_cols, axis=1), encoded_df], axis=1)

# Step 9: Check the final DataFrame
print(tips_encoded)
```

```python
# 2-> label encoding using sklearn

from sklearn.preprocessing import LabelEncoder

encoder=LabelEncoder()

df=pd.DataFrame({
    'color':['red','back','white','gray']
})
# fit and transform 
encoded_df=pd.DataFrame(encoder.fit_transform(df['color'])
)

print(f"{df} ")
print(f"{encoded_df}")

# here the vslus are provided on the aplhabetical basis 
# but to provide the values according to the rank 
# the ordinal encoding is used
```

```python
# ordinal encoding 

import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

df=pd.DataFrame({
    'size':['small','medium','large','medium','large']
})

encoder=OrdinalEncoder(categories=[['small','medium','large']])

encoder.fit_transform(df[['size']])

# encoder.transform([['small']])

```

```python
import pandas as pd
import seaborn as sns

# label encoding

df=pd.DataFrame({
    'city':['A', 'B', 'A', 'B', 'A','C'],
    'prices':[10, 20, 30, 40, 50,80]
})

mean_prices=df.groupby('city')['prices'].mean().to_dict()

df['mean_df']=df['city'].map(mean_prices)

df
```
