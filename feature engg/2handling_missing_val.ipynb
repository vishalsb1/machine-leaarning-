```python
import seaborn as sns

df=sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.isnull()
# it will return true for the column that have a null value
```

```python
df.isnull().sum()
# will eturn the true values(null) sum column wise
```

```python
df.dropna(axis=1)
# direct saglla udavlla NaN value aslela
```

```python
sns.histplot(df['age'],kde=True)
# etha je standar division davkavtay te jara veglay karna outliers present ahat
```

```python
# 2. mean impuataion

df["new_age"]=df['age'].fillna(df['age'].mean())

df['new_age'],df['age']
```

```python
sns.histplot(df['new_age'],kde=True)
```

```python
# atta samja outliers jasa atta ale tar vishay hou sakto so median wwala use kartat
```

```python
df['new_age1']=df['age'].fillna(df['age'].median())
```

```python
sns.histplot(df['new_age1'],kde=True)
```

```python
mode_val=df[df['age'].notna()]['age'].mode()[0]
mode_val
```

```python
# mode imputation

df['mode_age']=df['age'].fillna(mode_val)
```

```python
# df['age'].isnull().sum()
df['mode_age'].isnull().sum()
```

```python
sns.histplot(df['mode_age'])
```
