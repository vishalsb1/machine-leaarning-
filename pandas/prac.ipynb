# missing data


```python
data={
    'A':[1,2,np.nan,4,5],
    'B':[5,np.nan,np.nan,8,9],
    'C':[1,2,3,np.nan,np.nan],
    'D':[15,np.nan,np.nan,np.nan,19],
}
```

```python
df3=pd.DataFrame(data)
df3
```

```python
df3.isnull()
```

```python
df3.isnull().sum()
```

```python
df3.isnull().any() # it tells if there is any null value in the column  
```

```python
df3
```

```python
df3.dropna() # it drops the rows with any null value
```

```python
df3.dropna(thresh=2) # it drops the rows with less than 2 non-null values
```

```python
values={
    'A':0,
    'B':50,
    'C':33,
    'D':80   
}
```

```python
df3.fillna(value=values) # it fills the particular column with the given values
```

```python
df3.fillna(df3.mean()) # it fills the null values with the mean of the column
```

# mering joining and concatination...

```python
employees=pd.DataFrame({
    'employee_id':[1,2,3,4,5],
    'name':['john','anna','peter','linda','bob'],
    'department':['HR','Finance','IT','Marketing','Sales'],
})

salaries=pd.DataFrame({
    'employee_id':[1,2,3,4,5,6,7],
    'salary':[50000,60000,70000,80000,90000,100000,110000],
    'bonus':[5000,6000,7000,8000,9000,10000,11000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries,on='employee_id',how='inner') # inner join
```

```python
pd.merge(employees,salaries,on='employee_id',how='outer') 
```

```python
# concate mnje eka nanter durraa...

pd.concat([employees,salaries],axis=1) # it concates the two dataframes vertically
```

```python
# joining two dataframes with different columns

df1=pd.DataFrame({
    'name':['john','anna','peter','linda'],
},index=['a','b','c','d'])
df2=pd.DataFrame({
    'age':[24,13,53,33],
},index=['a','b','g','h'])

df1.join(df2,how='outer') # it joins the two dataframes with different columns
```
